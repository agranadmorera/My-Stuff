```python
# Importar librer√≠as para ejecutar ejercicios

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
olympic = pd.read_csv('athlete_events.csv')
```

```python
# Definir sexos a trav√©s de paleta de azules

sns.scatterplot(x='Age', y='Height', hue='Sex', data=olympic, palette='Blues')
plt.title('Age vs. Height')
```

```python
# diferenciar Females vs Males utilizando otro tipo de paleta, por ende, colores distintos

sns.scatterplot(x='Age', y='Height', hue='Sex', data=olympic, palette='PuOr')
plt.title('Age vs. Height')
```

```python
# Estatura de los atletas que compiten

sns.lineplot(x='Age', y='Height', data=olympic)
```

```python
# Estatura seg√∫n el sexo de los competidores (atletas)

sns.lineplot(x='Age', y='Height', data=olympic, hue='Sex')
```

```python
# Edad para cada atleta:

sns.lineplot(x='Age', y='Height', data=olympic, estimator=None)
```

```python
# Boxplot - edad de los atletas:

sns.boxplot(y=olympic['Age'])
plt.show()
```

```python
price = np.array([10, 45, 30, 15, 50])
std = np.array([2, 5, 8, 3, 1])

plt.xlabel("Stock")
plt.ylabel("Price")
plt.title("Stock Price")
plt.bar(x=[0,1,2,3,4], height=price, yerr = std,
        color="green", width=0.5)

plt.show()
```

```python
price = np.array([10, 45, 30, 15, 50])
std = np.array([2, 5, 8, 3, 1])

# Instead of [0,1,2,3,4], put the categories you need for each bar.
x_axis = ["APPLE", "AMAZON", "YAHOO", "GOOGLE", "FACEBOOK"]

plt.xlabel("Stock")
plt.ylabel("Price")
plt.title("Stock Price")
plt.bar(x=x_axis, height=price, yerr = std,
        color="teal", width=0.5)

plt.show()
```

```python
price = np.array([10, 45, 30, 15, 50])
std = np.array([2, 5, 8, 3, 1])
x_axis = ["APPLE", "AMAZON", "YAHOO", "GOOGLE", "FACEBOOK"]

plt.xlabel("Stock")
plt.ylabel("Price")
plt.title("Stock Price")
plt.bar(x=x_axis, height=price, yerr = std,
        color="blue", width=0.5)

plt.grid(True)
plt.show()
```

```python
# Del Dataset bitcoin se extraen estos gr√°ficos en Matplotlib

bitcoin = pd.read_csv("Bitcoin.csv")
bitcoin.head()
```

```python
fig=plt.figure(figsize=(12,8))

plt.grid(True)
plt.plot(bitcoin.Date, bitcoin.Price)
plt.xticks(rotation=30)

for x, y in zip(bitcoin.Date, bitcoin.Price):
    plt.text(x, y, y, ha='center', va='bottom')
```

```python
fig=plt.figure(figsize=(15,8))

plt.xticks(rotation=30)
plt.grid(True)
plt.bar(bitcoin.Date, bitcoin.Price)

for x, y in zip(bitcoin.Date, bitcoin.Price):
    plt.text(x, y, y, ha='center', va='bottom')
```
